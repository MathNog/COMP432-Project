## ARIMA Model

### Description:

This notebook will implement code for time series analysis and forecasting with ARIMA models.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

### Some useful functions


```python
#Ploting a single time series

def plot_series(serie,title,xlabel,ylabel):
    plt.figure()
    plt.plot(serie)
    plt.title=title
    plt.xlabel=xlabel
    plt.ylabel=ylabel
    plt.show()
    
#Compute some error metrics

def errors(actual, predicted):
    mape = np.mean(np.abs(predicted - actual)/np.abs(actual))  # MAPE
    me = np.mean(predicted - actual)             # ME
    mae = np.mean(np.abs(predicted - actual))    # MAE
    mpe = np.mean((predicted - actual)/actual)   # MPE
    rmse = np.mean((predicted - actual)**2)**.5  # RMSE
    
    errors = [['MAPE',mape],['ME',me],['MAE',mae],['MPE',mpe],['RMSE',rmse]]
    
    errors = pd.DataFrame(errors,columns=['Error','Value'])
    return errors
```

```python
#Importing the dataset and making some changes

series = pd.read_csv('../data/AAL2010_2019.csv')
series = series.loc[:,["Date","Adj Close"]]
series.set_index("Date",inplace=True)
series.index=pd.to_datetime(series.index)
series = series.asfreq('d')
# series.index = pd.DatetimeIndex(series.index)

series.head()
```

## Coments regarding the series behavior

By seeing the series it seems non-stationary and non-seasonal.
In order to verify our assumption, we will pass the serie through a stationarity test and a seasonal decomposition

```python
adf = adfuller(series)
print(adf)
```

We have 2 reasons to differenciate the series:
 - It is not stationary
 - By doing the first difference we will have the stock *return*, which is usually more used than the price itself

```python
series_diff1 = series - series.shift(1)
series_diff1 = series_diff1.dropna() #the first value of diff will be NaN, so we have to drop it
plot_series(series_diff1,"American Airlines Adj Close Return- Monthly",xlabel="Time",ylabel="Return")
```

```python
adf = adfuller(series_diff1)
print(adf)
```

### Conclusion regarding stationarity

The second value from the variable adf is the p-value, which tells if the series is stationary or not.
Since it is bigger than 0.05, the **Null Hypothesis** is not rejected, hence the series is **not stationary**.

We have differenciated the series in order to get a stationary series, even though the ARIMA Model could have taken care of this, since the *I* in *ARIMA* means *Integrated*, so this model is able the differenciate a Time Series if needed. One differentiation was enought to have a p-values smaller than 0.05

```python
seasonal_result = seasonal_decompose(series_diff1,period=12)
seasonal_result.plot()
```

### Conclusion regarding Seasonality

As observed in the plot, our time series do not have a seasonal behavior.

We can also see that the trend doe not have a crescent or decrescent behavior, which, once more, confirms the series stationary behavior.

With that bieng said, there is nothing to do with the series. 

## Correlation Analysis

The next step in out Time Series Analysis should be to create the **Autocorrelation and Partial Autocorrelation Functions**, in order to determine the parameters of the **ARIMA Model**.

### ARIMA Model Brief Introduction

The fundamentals of ARIMA Models are, actually, two other Time Series Models, the AR and MA.

The **AR**, Autoregressive Model, takes the past values of the series to predict the future ones. It relies on the assumption that the present values is a linear combination of lagged values.

The **MA**, Moving Average Model, relies on the assumption that the regression errors are a linear combination of the errors terms from past values.

The **ARIMA** model combines those two models and, also, is capable of *differenciate* the series to get a stationary behavior, since this is an assumption of the model as well.

In order to determine the parameters of the model, the past values that will be used to predict the present and future ones, we use the Autocorrelation and Partialcorrelation functions.

As we can see, this model is quite restrictive, even though is a classic model. We aim to see if Forecasting with Neural Networks will be a more flexible and precise approach.

```python
plot_acf(series_diff1)
```

```python
plot_pacf(series_diff1)
```

### Parameters chosen

Since the significant lags are, for both functions 1, we will chose the orders to be (1,0,1).

Note that the *d* parameter is zero because we have already differentiated the series - we are dealing ith a return series.

## Training and Test Sets

We must divide our time series into a training and test sample. Since our series has anual seasonality, let's leave the last yeas as test sample and the rest of the series as training.

```python
train_series = series_diff1.iloc[0:len(series_diff1)-365]
test_series = series_diff1.iloc[len(series_diff1)-365:]

print(len(train_series)+len(test_series),len(series_diff1))
```

```python
model_arima = ARIMA(train_series,order=(1,0,1)).fit()
```

```python
print(model_arima.summary())
```

```python
# Plot residual errors
residuals = pd.DataFrame(model_arima.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title="Residuals", ax=ax[0])
residuals.plot(kind='kde', title='Density', ax=ax[1])
plt.show()
```

Our residuals density are fairly normal with mean zero and variance constant

### Prediction

Once your model is fitted, we can predict the value of the series with the model and compare the results.

```python
prediction_arima = model_arima.predict()

compare_prediction = train_series.copy(deep=True)
compare_prediction['Prediction'] = prediction_arima
plt.plot(compare_prediction);
```

### Model conclusion

By looking at the plot above, we see that the **ARIMA** Model does not predict the correct behavior of the series. All of its predictions are extremely centered in the mean of the series, while the actual values cover a larger area.

Even though the model is not good, lets compute the error between the prediction and the actual values and, after that, make the forecast for 1 whole year.

```python
prediction_errors = errors(compare_prediction['Adj Close'],compare_prediction['Prediction'])

prediction_errors
```

### Forecasting

```python
forecast_arima = model_arima.predict(start='2018-12-31',end='2019-12-30')

compare_forecast = test_series.copy(deep=True)
compare_forecast['Forecast'] = forecast_arima

plt.plot(compare_forecast);
```

We can see that the forecast is quite horrible, actually. If we look at the forecast values, we can see that after a few steps it stays always in the same value, the mean

```python
forecast_arima
```

```python
Forecast_errors = errors(compare_forecast['Adj Close'],compare_forecast['Forecast'])

Forecast_errors
```
